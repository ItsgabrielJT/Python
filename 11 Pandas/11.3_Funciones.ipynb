### Combinacion de dataframes


```python
daf1 = pd.DataFrame(
    {
        'A': ["A01", "A02", "A03", "A04"],
        'B': ["B01", "B02", "B03", "B04"],
        'C': ["C01", "C02", "C03", "C04"],
        'D': ["D01", "D02", "D03", "D04"]
    }
)

daf2 = pd.DataFrame(
    {
        'A': ["A05", "A06", "A07", "A08"],
        'B': ["B05", "B06", "B07", "B08"],
        'C': ["C05", "C06", "C07", "C08"],
        'D': ["D05", "D06", "D07", "D08"]
    }
)
```

### Concat

```python
# Concat, permite combinar tablas por filas o columnas
# Por default se agrupa por filas y el indice se repite

pd.concat([daf1, daf2])
```

```python
# Para eviatr que se repitan los indices

pd.concat([daf1, daf2], ignore_index=True)
```

```python
# Para hacer que se agrupen por columnas

pd.concat([daf1, daf2], axis=1)
```

### Merge

```python
izq = pd.DataFrame(
    {
        'key': ['k1', 'k2', 'k3', 'k4'],
        'A': ['A1', 'A2', 'A3', 'A4'],
        'B': ['B1', 'B2', 'B3', 'B4'],
    }
)

der = pd.DataFrame(
    {
        'key': ['k1', 'k2', 'k3', 'k4'],
        'C': ['C1', 'C2', 'C3', 'C4'],
        'D': ['D1', 'D2', 'D3', 'D4']
    }
)
```

```python
# Merge combina la tablas por un elemtno en comun, aqui le estamos diciendo que el elemento comun es key
# De no tener un elemtno en comun nos lanzara una excepcion 
izq.merge(der, on='key')
```

### Join
Trabaja directamente con los indices y ya no con las columnas de cada dataframe como lo hacia merge

#### Left Join
Mantiene todos los datos de la tabla izquierda y trae solo los datos que coincidan de la tabla derecha

#### Right Join 
Mantiene todos los datos de la tabla derecha y trae solo los datos que coincidan de la tabla izquierda

#### Inner Join 
Atre los datos que coincidan en ambas tablas

#### Outer Join 
Mantiene los daros que coinciden en las tablas y atrae toda la informacion de ambas tablas

```python
join1 = pd.DataFrame(
    {
        'A': ["A1","A2","A3"],
        'B': ["B1","B2","B3"]
    },
    index=[1, 2, 3]
)

join2 = pd.DataFrame(
    {
        'C': ["C1","C2","C3"],
        'D': ["D1","D2","D3"]
    },
    index=[1, 4, 3]
)

```

```python
# Join coincicido tablas a partir de un inidice en comun, por defualt nos hace un left join
join1.join(join2)
```

```python
# Si queremos acer otros tipos de join
join1.join(join2, how='inner')
```
